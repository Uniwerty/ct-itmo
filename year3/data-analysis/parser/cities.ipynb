```python
cities2 = cities.loc[cities.Elevation > 0]
plt.hist(cities.Household_income, bins=200, log=False)
```

```python
cities3 = cities2.loc[cities2.Longitude > -125]
plt.figure(figsize=(10, 5))
plt.scatter(cities3.Longitude, cities3.Latitude, c=cities3.Elevation, cmap='turbo', s=0.3)
plt.colorbar(norm=cities3.Elevation, cmap='turbo')
plt.title('US cities elevation')
plt.savefig('US_elevation.png')
```

```python
cities4 = cities.loc[cities.Density < 10000]
cities4 = cities4.loc[cities4.Longitude > -125]
plt.figure(figsize=(10, 5))
plt.scatter(cities4.Longitude, cities4.Latitude, c=cities4.Density, cmap='Reds', s=0.3)
plt.colorbar(norm=cities4.Density, cmap='Reds')
plt.title('US cities density')
plt.savefig('US_density.png')
```
